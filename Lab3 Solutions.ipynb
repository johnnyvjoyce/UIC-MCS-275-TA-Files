```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
countries = ["USA", "Canada", "China", "UK", "Russia", "Italy"]
cities = [["Chicago", "Boston", "San Francisco"],["Ottawa","Vancouver","Montreal"],["Beijing","Shanghai","Shenzhen"],["London","Bristol","Cardiff"],["Moscow","St Petersburg","Vladivostok"],["Rome", "Florence", "Naples"]]

class CountryClass():
    def __init__(self, country, cities):
        self.countryname = country
        self.city1 = cities[0]
        self.city2 = cities[1]
        self.city3 = cities[2]
        
countriesandcities = []
for country, cities in zip(countries,cities):
    countriesandcities.append(CountryClass(country, cities))

print([i.__dict__ for i in countriesandcities])
```

```python
def Question6(n: int):
    fig = plt.figure()
    ax = plt.axes(projection="3d")
    randomxvalues = np.random.random(n)
    randomyvalues = np.random.random(n)
    zvalues = [np.exp(-(x**2+y**2)) for (x,y) in zip(randomxvalues,randomyvalues)]
    ax.scatter(randomxvalues, randomyvalues,zvalues)
    
    
Question6(n=500)
```

```python
def Question10(box_size: int):

    linestring = ""
    for i in range(4): # Want 4 dark parts with 4 blank parts separating them
        linestring += "#"*box_size
        linestring += " "*box_size
    reverselinestring = linestring[::-1] # Why does [::-1] reverse a string?
    # reverselinestring will be for the rows where the blank spaces come before the dark spaces

    checkerstring = "" # This will be a string for 1 "row" of the checkerboard
    reversecheckerstring = ""
    for i in range(box_size):
        checkerstring += linestring + "\n" # "\n" denotes a new line
        reversecheckerstring += reverselinestring + "\n"

    checkerboard = (checkerstring + reversecheckerstring)*2 # Create the final board which we'll output

    newlist = []
    sublist = []
    for i in checkerboard:
        if i == "\n":
            newlist.append(sublist)
            sublist = []
        elif i == "#":
            sublist.append(1)
        elif i == " ":
            sublist.append(0)

    checkerlist = np.array(newlist)
    plt.imshow(checkerlist,cmap="gray")
    
Question10(box_size=1)
```
