```python
###################
### Question 3 ####
###################

def Q3(a, b):
    try:
        a = float(a)
        b = float(b)
        result = a**b
        assert type(result) != complex
        return result
    except OverflowError:
        return "infinity"
    except ValueError:
        return "error"
    except AssertionError:
        return "complex number"
    
print(Q3(3,2))
print(Q3("three","two")) # "error"
print(Q3(999,999)) # "infinity"
print(Q3(-1,0.5)) # "complex number"
```

```python
###################
### Question 4 ####
###################

class Node(object):
        def __init__(self, value = None):
            """
            basic structure of a binary tree is a single node.
            a node can hold a node value and it
            lays out a connection for other nodes
            to be placed either left or right during the next recursive
            pass of the binary tree construction process. 
            initially the node is empty (i.e. value = None) but it 
            posesses all features to become a tree. 
            all values that are used during the construction of the 
            binary tree are contained in a list L.
            the first element in list L, L[0], becomes the value of the
            root node. the rest of the nodes are then added to it using the
            insert_node() method. see construct_binary_tree(L) method.
            """
            self.value = value
            self.left = None
            self.mid = None
            self.right= None

        def insert_node(self, new_value):
            """
            this method inserts a new node into the object
            recursively, using the standard convention:
            values equal or smaller than the root go to the left,
            the rest goes to the right.
            """
            if new_value < self.value:
                if self.left == None:
                    self.left = Node(new_value)
                else:
                    # This is the recursive function call
                    self.left.insert_node(new_value)
                    
            elif new_value == self.value: 
                if self.mid == None:
                    self.mid = Node(new_value)
                else:
                    self.mid.insert_node(new_value)

            else: #case when new_value > self.value:
                    if self.right == None:
                        self.right = Node(new_value)
                    else:
                        self.right.insert_node(new_value)

        def construct_ternary_tree(self, L):
        # L is a list of values that will be in the tree, in the order that they will be inserted
            self.value = L[0]
            for ele in L[1:]:
                self.insert_node(ele)

        def print_tree(self):
            print(self.value)
            if self.left == None:
                print("left: none!")
            else:
                print("left: ", self.left.value)
            
            if self.mid == None:
                print("mid: none!")
            else:
                print("mid: ", self.mid.value)
            
            if self.right == None:
                print("right: none!")
            else:
                print("right: ", self.right.value)



        # This will print all of the elements in the level "lvl" from left to right.
        def print_lvl(self, lvl):
            # lvl = 0 means the root level, lvl = 1 means the level below the root, etc.
            if lvl == 0:
                print(self.value)
            if self.left != None:
                self.left.print_lvl(lvl-1)
            if self.mid != None:
                self.mid.print_lvl(lvl-1)
            if self.right != None:
                self.right.print_lvl(lvl-1)


        def inorder_traversal(self):
            if self.left != None: #alternatively, if self.left: 
                self.left.inorder_traversal()  #go a level deeper
            print(self.value)
            if self.mid != None:
                self.mid.inorder_traversal()
            if self.right != None: #if node on right is not None!
                self.right.inorder_traversal() #go a level deeper
        


        # Printing the values in the tree in a bottom-up and left to right fashion
        def postorder_traversal(self):
            if self.left != None: #alternatively, if self.left: 
                self.left.postorder_traversal()  #go a level deeper
            if self.mid != None:
                self.mid.postorder_traversal()
            if self.right != None: #if node on right is not None!
                self.right.postorder_traversal() #go a level deeper
            print(self.value)



        # Printing the values in the tree in a sort of top to bottom and left to right fashion
        def preorder_traversal(self):
            print(self.value)
            if self.left != None: #alternatively, if self.left: 
                self.left.preorder_traversal()  #go a level deeper
            if self.mid != None:
                self.mid.preorder_traversal()
            if self.right != None: #if node on right is not None!
                self.right.preorder_traversal() #go a level deeper
                
        # Gets the depth of the binary tree (distance of the farthest node from the start)
        def depth(self):
            if self.left == self.right == self.mid == None:
                return 0
            leftdepth, middepth, rightdepth = 0, 0, 0 
            if self.left != None:
                leftdepth = self.left.depth()
            if self.mid != None:
                middepth = self.mid.depth()
            if self.right != None:
                right = self.right.depth()
            return max(leftdepth+1,middepth+1,rightdepth+1)

        # Counts the number of nodes with degree d
        def countdegree(self, d):
            degree = 0
            leftcount = 0 # Number of times we find degree d on left subtree
            midcount = 0 # Number of times we find degree d on middle subtree
            rightcount = 0  # Number of times we find degree d on right subtree
            if self.left != None:
                degree += 1
                leftcount = self.left.countdegree(d)
            if self.mid != None:
                degree += 1
                midcount = self.mid.countdegree(d)
            if self.right != None:
                degree += 1
                rightcount = self.right.countdegree(d)
            
            if degree == d: # If the degree of the current node is d, add 1 to the overall count
                return leftcount + midcount + rightcount + 1
            else: # Else, just count the degrees of the subtrees
                return leftcount + midcount + rightcount

# new_tree = Node()
# new_tree.construct_ternary_tree([5, 1, 0, 2, 6, 8, 7, 9])
# new_tree.print_lvl(2)

new_tree = Node()
new_tree.construct_ternary_tree([5,7,3,5,4])
print("Level 0:")
new_tree.print_lvl(0)
print("Level 1:")
new_tree.print_lvl(1)
print("Level 2:")
new_tree.print_lvl(2)

print("\n\nDepth:")
print(new_tree.depth())


print("Number of nodes with degree 0:")
print(new_tree.countdegree(0))


print("Number of nodes with degree 1:")
print(new_tree.countdegree(1))

print("Number of nodes with degree 2:")
print(new_tree.countdegree(2))

print("Number of nodes with degree 3:")
print(new_tree.countdegree(3))
```

```python
###################
### Question 5 ####
###################

def printboard(L):
    #print("-------------")
    for i in range(3): # print each row
        print("| %s | %s | %s | " % (L[i][0],L[i][1],L[i][2]))
        if i < 2:
            print("+---+---+---+")
        
        
        
def checkvictory(L, player):
    '''
    Check if the given player has won in game state L
    '''
    for i in range(3): # Check if player has all three in a single row
        if L[i][0] == L[i][1] == L[i][2] == player:
            return True
        
    for i in range(3): # Check if player has all three in a single column
        if L[0][i] == L[1][i] == L[2][i] == player:
            return True
        
    if L[0][0] == L[1][1] == L[2][2] == player: # Check the first diagonal
        return True
    if L[2][0] == L[1][1] == L[0][2] == player: # Check the second diagonal
        return True
    
    return False # If none of the check have returned true, then return false


def getpossiblemoves(L):
    '''
    Returns a list of tuples representing the coordinates of each blank space
    '''
    blanks = []
    for i in range(3):
        for j in range(3):
            if L[i][j] == " ":
                blanks.append((i,j))
    return blanks
    
        
        
myboard = [ [" ","x","o"], ["x", " ", "o"], [" ", "x", "o"] ]

printboard(myboard)

print("checkvictory(myboard, \"x\"): ", checkvictory(myboard, "x"))

print("checkvictory(myboard, \"o\"): ", checkvictory(myboard, "o"))

print("getpossiblemoves(myboard) : ", getpossiblemoves(myboard))
```
