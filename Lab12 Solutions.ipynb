```python
###################
### Question 1 ####
###################


import urllib3
from bs4 import BeautifulSoup

def Weather():
    
    http = urllib3.PoolManager() 
    agent = {"User-Agent":'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'}
    url = "https://forecast.weather.gov/MapClick.php?lat=41.883210000000076&lon=-87.65876999999995"
    requested = http.request('GET', url,headers=agent)
    B = BeautifulSoup(requested.data.decode('utf-8'), "html5lib") # html5lib is the actual parser that processes HTML/XML documents

    # print(B) # for a basic HTML content
    # print(B.prettify()) # for a better/neater display of the HTML content

    days = B.find_all('div',{"class":"tombstone-container"}) # tombstone-container is the name of the div we want on the page

    for timeperiod in list(days): # Go through each day
        
        ''' First get the name of the time period and strip the html tags from it '''
        name = str(timeperiod.find("p")) # Get the name of the time period
        name = name.replace('<p class="period-name">','') # Remove tags, part 1
        name = name.replace('</p>','') # Remove tags, part 2
        name = name.replace('<br/>',' ') # Remove tags, part 3
        
        ''' Then get the corresponding temperature and strip the html tags'''
        temperature = str(timeperiod.find("p", {"class":"temp temp-high"}))
        temperature = temperature.replace('<p class="temp temp-high">','') # Remove tags, part 1
        temperature = temperature.replace('</p>','') # Remove tags, part 2
        
        ''' If the time period is a night, then look for the low temperature instead'''
        if temperature == "None":
            temperature = str(timeperiod.find("p", {"class":"temp temp-low"}))
            temperature = temperature.replace('<p class="temp temp-low">','') # Remove tags, part 1
            temperature = temperature.replace('</p>','') # Remove tags, part 2
        
        print(name)
        print(temperature)
        print("\n")


Weather()
```

```python
import urllib3
from bs4 import BeautifulSoup

def Google(query = "university+of+illinois+at+chicago"):
    # PoolManager class automatically handles creation of ConnectionPool instances for each host. 
    # By default, number of instance it can handle is 10. This number can be increased. 
    http = urllib3.PoolManager() 
    query = query.replace(" ", "+") # Replace spaces with plus symbols so it works in the url
    search_url = "https://www.google.com/search?q=%s" % query # Put the query into the search bar
    requested = http.request('GET', search_url)
    B = BeautifulSoup(requested.data, "html5lib") # html5lib is the actual parser that processes HTML/XML documents

    searchresults = B.find_all('div',attrs={"class":"kCrYT"}) 
    # This "kCyRT" class was found by downloading and rendering the page, then using inspect element

    for result in searchresults: # Go through each result
        
        try: # Get the link
            link = result.find("a")["href"]
            link = str(link).replace("/url?q=","")
        except TypeError: # if link leads nowhere
            link = "None"

        
        name = str(result.find("div",attrs={"class":"BNeawe s3v9rd AP7Wnd"})) 
        # The "BNeawe s3v9rd AP7Wnd" was found by downloading and rendering the HTML, then using inspect element
        name = name.replace('<div class="BNeawe s3v9rd AP7Wnd">',"") # remove the THML tags
        name = name.replace("</div>","")
        name = name.replace("<div>","")
        
        print(link)
        print(name)
        print("\n")

    '''NOTE:
        Some of these urls don't align with the results. This is because this Google page is non-uniform and includes some
        info about UIC and some results in the same html tags. This can be fixed with more fine-tuning, but this example
        gives the main idea!
    '''

Google()
```

```python
###################
### Question 3 ####
###################

import urllib3
from bs4 import BeautifulSoup

def Flight(flightcode = "LON-PAR"): # Default flight route: London to Paris
    # PoolManager class automatically handles creation of ConnectionPool instances for each host. 
    # By default, number of instance it can handle is 10. This number can be increased. 
    http = urllib3.PoolManager() 
    url = "https://www.kayak.com/flights/%s/2021-05-08/2021-05-15?sort=bestflight_a" % flightcode # Kayak website
    requested = http.request('GET', url)
    B = BeautifulSoup(requested.data.decode("utf-8"), "html5lib") # html5lib is the actual parser that processes HTML/XML documents

    flights = B.find_all('div', attrs={"class":"multibook-dropdown"}) # Separate each result in the flight search

    for flight in list(flights): # Iterate over each flight result
        
        airline = str(flight.find("span", attrs={"class":"name-only-text"})) # Get the name of the airline
        airline = airline.replace('<span class="name-only-text">','') # Remove HTML tags
        airline = airline.replace('</span>','') # Remove HTML tags
        airline = airline.replace('\n','') # Remove newline characters
        if "<" in airline: # Some results have more HTML tags inside them other than the "span" ones. We can use this to cut off the remaining html
            airline = airline[:airline.index("<")] # Delete everything after the opening HTML tag
        
        price = str(flight.find('span', attrs={"class":"price-text"})) # Get the price of the trip
        price = price.replace('</span>','') # Remove HTML tags
        price = price[price.index("$"):] # Ignore all characters in front of "$" symbol (because this is an HTML tag with the ID, but it is different for each result)
        price = price.replace("\n", '') # Remove newline characters
        

        print("%s (%s)" % (airline, price))



Flight()
```
